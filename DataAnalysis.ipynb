```python
price_ranges = pd.cut(shuffled_df['price/value'], bins=[0, 1000, 2500, 5000,7500,10000,15000,20000,30000,50000, float('inf')], labels=['<1k', '1k-2.5k', '2.5k-5k','5k-7.5k','7.5k-10k','10k-15k','15k-20k','20k-30k','30k-50k','>50k'])
price_ranges.value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Price Range Distribution in INR')
plt.show()
```

```python
count=2
for count in range (1,14,4):
    top_brands = shuffled_df.groupby('brands').filter(lambda x: (x['stars'].mean() >= 4.0) & (x['reviewsCount'].sum() > 2000*count))

    # Get the top 20 brands based on average price
    top_20_brands = top_brands.groupby('brands')['stars'].mean().nlargest(20).index

    # Filter the DataFrame for the top 20 brands
    top_20_brands_df = top_brands[top_brands['brands'].isin(top_20_brands)]

    # Calculate average price for each top 20 brand
    avg_price_by_brand = top_20_brands_df.groupby('brands')['price/value'].mean()

    # Plot the bar plot for top 20 brands and their average price
    plt.figure(figsize=(14, 8))
    barplot=sns.barplot(x=avg_price_by_brand.index, y=avg_price_by_brand.values, palette='viridis')
    plt.title(f'Average Price of Products for Top 20 Brands (Stars >= 4.0, Reviews > {2000*count})')
    plt.xlabel('Brand')
    plt.ylabel('Average Price')
    plt.xticks(rotation=45, ha='right')
    for index, value in enumerate(avg_price_by_brand.values):
        barplot.text(index, value + 100, f'{value:.2f}', ha='center', va='bottom',rotation=30, fontsize=8, color='black')
    plt.show()
```
