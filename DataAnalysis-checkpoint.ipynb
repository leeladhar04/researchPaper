```python
import pandas as pd 
import os 
import numpy as np
```

```python
path=os.getcwd()

csv_files = [f for f in os.listdir(path) if f.endswith('.csv')]
print(csv_files)
```

```python

processedFrame=[]
for i in csv_files:
    df=pd.read_csv(i)
    df=df.drop(columns=["title","price/currency","description","asin","thumbnailImage","breadCrumbs","url"])
    df=df.dropna()
    df=df.drop_duplicates()
    processedFrame.append(df)
    
```

```python
concatenated_df=pd.concat(processedFrame,ignore_index=True)
```
